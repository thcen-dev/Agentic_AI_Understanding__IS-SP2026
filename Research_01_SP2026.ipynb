## 1. â“What is an ADK (Agentic Development Kit)?



###Definition

To put it simply, an `ADK` is essentially a toolkit that helps us build AI agents. `AI agents` are programs that can reason, take actions, use tools, and interact with users and or systems independently.

###What is "Agentic"?

When using the term `Agentic`, it refers to the AI being able to act with a distinct level oif independence rather than responding to a prompt-to-prompt scenario; essentially it is an agent working for you in some capacity. This essentially allows an AI agent to handle more complex and multi-step workflows.

##2. âœ…âŒPros & Cons (Gemini vs OpenAI)



###Gemini Pros & Cons

The Gemini ADK works best at `structured agent workflows` which is a sequence of predefined roles, steps, and decisions that help it in its overall decision-making. Having this structured, coded-out structure allows for the agent to manage complex systems in a very relaible manner. It also integrates very well with Google Cloud and Vertex AI.

Where Gemini lacks however is its intuitive learning curve, as it can be harder to use given its ecosystem. The structure of Gemini ADK can also limit flexibility in regard to working with more dynamic or creative agent behaviors. The setup and reliance on Google-managed infrastructure can also introduce unnecessary complexity for small scale projects.



###OpenAI Pros & Cons

The OpenAI ADK works best at flexible and dynamic agent interactions, allowing developers to rapidly prototype agents that can reason, call tools, and adapt to a wide range of tasks. Its developer-friendly APIs and extensive documentation make it intuitive to use, while strong community support enables faster troubleshooting and experimentation.

Where OpenAI lacks however is in its level of built-in structure for managing complex, multi-agent workflows at scale. This flexibility can sometimes lead to less predictable behavior without additional guardrails or custom logic. Additionally, advanced agent orchestration and long-running autonomy often require more manual configuration by the developer.

###Editing the Agent

After creating and integrating my API key, I then began to edit the contents of my agent, in which I entered this code into the `agent.py` file:



```
from google.adk.agents.llm_agent import Agent

# Mock tool implementation
def get_current_time(city: str) -> dict:
    """Returns the current time in a specified city."""
    return {"status": "success", "city": city, "time": "10:30 AM"}

root_agent = Agent(
    model='gemini-3-flash-preview',
    name='root_agent',
    description="Tells the current time in a specified city.",
    instruction="You are a helpful assistant that tells the current time in cities. Use the 'get_current_time' tool for this purpose.",
    tools=[get_current_time],
)
```

This specific edit is responsibe for making an interactive agent that is able to respond to prompting in regard to telling you the specifc time of a city at my request.
